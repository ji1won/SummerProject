## data.yaml 경로 설정

```python
from glob import glob

# 이미지 경로 list로 넣기
train_img_list = glob('/content/drive/MyDrive/yolo_sample/train/images/*.jpg') + glob('/content/drive/MyDrive/yolo_sample/train/images/*.jpeg')
valid_img_list = glob('/content/drive/MyDrive/yolo_sample/valid/images/*.jpg') + glob('/content/drive/MyDrive/yolo_sample/valid/images/*.jpeg')
test_img_list = glob('/content/drive/MyDrive/yolo_sample/test/images/*.jpg') + glob('/content/drive/MyDrive/yolo_sample/test/images/*.jpeg')


# txt 파일에 write
with open('/content/drive/MyDrive/yolo_sample/train.txt', 'w') as f:
	f.write('\n'.join(train_img_list) + '\n')

with open('/content/drive/MyDrive/yolo_sample/valid.txt', 'w') as f:
	f.write('\n'.join(valid_img_list) + '\n')

with open('/content/drive/MyDrive/yolo_sample/test.txt', 'w') as f:
	f.write('\n'.join(test_img_list) + '\n')
```

```python
import yaml
with open('/content/drive/MyDrive/yolo_sample/data.yaml', 'r') as f:
  data = yaml.load(f, Loader=yaml.FullLoader)
print(data)
data['names'] = {"old2", "young2"}
data['train'] = '/content/drive/MyDrive/yolo_sample/train.txt'
data['val'] = '/content/drive/MyDrive/yolo_sample/valid.txt'
data['test'] = '/content/drive/MyDrive/yolo_sample/test.txt'

with open('/content/drive/MyDrive/yolo_sample/data.yaml', 'w') as f:
  yaml.dump(data, f)
#print(data)
```

## model train
 - --img : input image size
 - --batch : batch size
 - --epochs : epochs 훈련 횟수
 - --data : yaml 경로
 - --cfg : pretrained model (ex. yolov5m.yaml)
 - --weights : pretrained model weight (ex. yolov5m.pt)
 - --name : 학습정보 저장 폴더 이름 (yolov5/runs/train/이름)
